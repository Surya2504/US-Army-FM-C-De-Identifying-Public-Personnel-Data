```python
# importing the necessary modules
import requests
from bs4 import BeautifulSoup
from zipfile import ZipFile
from io import BytesIO
import pandas as pd
import os
from IPython.display import display
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Set variable for page to be opened and url to be concatenated 
page =requests.get('https://www.opm.gov/data/index.aspx')
baseurl= 'https://www.opm.gov/'
#Use BeautifulSoup to clean up the page
soup = BeautifulSoup(page.text, "html.parser")
soup.prettify()

# Creating a new file to store the zip file links
newfile = open('zipfiles.txt','w')
#Find all the links on the page that end in .zip and write them into the text file
for anchor in soup.findAll('a', href=True):
    links = anchor['href']
    if links.endswith('.zip'):
        newfile.write(links + '\n')
newfile.close()
```

```python
#Fetching the links for the zip file and downloading the files
with open('zipfiles.txt', 'r') as links:
    count = 0
    for link in links:
        if count == 2:
            break
        if link:
            filename1= link.split('/')[-1]
            filename= filename1[:-1]
            link = baseurl + link
            print(filename + ' file started to download')
            response = requests.get(link[:-1])
            # Writing the zip file into local file system
            with open(filename,'wb') as output_file:
                output_file.write(response.content)
            print(filename + ' file is downloaded')
            with ZipFile(filename, 'r') as zip:
                # Get Agency Translation data
                with zip.open('DTagy.txt', 'r') as myfile:
                    DTagy = myfile.read().decode("utf-8")
                    #print(DTagy)
                # Get Location Translation data
                with zip.open('DTloc.txt', 'r') as myfile:
                    DTloc = myfile.read().decode("utf-8")
                    #print(DTloc)
                # Get Age Translation data
                with zip.open('DTagelvl.txt', 'r') as myfile:
                    DTagelvl = myfile.read().decode("utf-8")
                    #print(DTagelvl)
                # Get Education Level Translation data
                with zip.open('DTedlvl.txt', 'r') as myfile:
                    DTedlvl = myfile.read().decode("utf-8")
                    #print(DTedlvl)
                # Get General Schedule & Equivalent Grade Translation data
                with zip.open('DTgsegrd.txt', 'r') as myfile:
                    DTgsegrd = myfile.read().decode("utf-8")
                    #print(DTgsegrd)
                # Get Length of Service Translation data
                with zip.open('DTloslvl.txt', 'r') as myfile:
                    DTloslvl = myfile.read().decode("utf-8")
                    #print(DTloslvl)
                # Get Occupation Translation data
                with zip.open('DTocc.txt', 'r') as myfile:
                    DTocc = myfile.read().decode("utf-8")
                    #print(DTocc)
                # Get Occupation Category Translation data
                with zip.open('DTpatco.txt', 'r') as myfile:
                    DTpatco = myfile.read().decode("utf-8")
                    #print(DTpatco)
                # Get Pay Plan & Grade Translation data
                with zip.open('DTppgrd.txt', 'r') as myfile:
                    DTppgrd = myfile.read().decode("utf-8")
                    #print(DTppgrd)
                # Get Salary Level Translation data
                with zip.open('DTsallvl.txt', 'r') as myfile:
                    DTsallvl = myfile.read().decode("utf-8")
                    #print(DTsallvl)
                # Get STEM Occupations Translation data
                with zip.open('DTstemocc.txt', 'r') as myfile:
                    DTstemocc = myfile.read().decode("utf-8")
                    #print(DTstemocc)
                # Get Supervisory Status Translation data
                with zip.open('DTsuper.txt', 'r') as myfile:
                    DTsuper = myfile.read().decode("utf-8")
                    #print(DTsuper)
                # Get Type of Appointment Translation data
                with zip.open('DTtoa.txt', 'r') as myfile:
                    DTtoa = myfile.read().decode("utf-8")
                    #print(DTtoa)
                # Get Work Schedule Translation data
                with zip.open('DTwrksch.txt', 'r') as myfile:
                    DTwrksch = myfile.read().decode("utf-8")
                    #print(DTwrksch)
                # Get Work Status Translation data
                with zip.open('DTwkstat.txt', 'r') as myfile:
                    DTwkstat = myfile.read().decode("utf-8")
                    #print(DTwkstat)
                # Get Date Translation data
                with zip.open('DTdate.txt', 'r') as myfile:
                    DTdate = myfile.read().decode("utf-8")
                    #print(DTdate)
                
                z = ZipFile(BytesIO(response.content))
            # gets September data
            if count == 0:
                def new_directory(directory, filename):
                  # Before creating a new directory, check to see if it already exists
                  if os.path.isdir(directory) == False:
                    os.mkdir(directory)
                #print(directory+" "+filename)
                z.extractall('SepData')
                print('September data Extracted')
                print('-------------------------------------------------------------')
            # gets June data    
            elif count == 1:
                def new_directory(directory, filename):
                  # Before creating a new directory, check to see if it already exists
                  if os.path.isdir(directory) == False:
                    os.mkdir('JunData')
                z.extractall('JunData')
                print('June data Extracted')
            count += 1
                
```
